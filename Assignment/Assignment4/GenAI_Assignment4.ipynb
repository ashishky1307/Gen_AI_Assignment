```python
from transformers import pipeline

# Initialize the QA pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

def ask_question(question):
    context = (
        "Chatbots are AI-based systems that simulate a conversation with users. "
        "They are used in customer service and other applications to respond to queries."
    )
    result = qa_pipeline(question=question, context=context)
    return result['answer']

# Get user input for the question
user_question = input("Ask a question about chatbots: ")
answer = ask_question(user_question)

print("\nQuestion:", user_question)
print("Answer:", answer)
```
