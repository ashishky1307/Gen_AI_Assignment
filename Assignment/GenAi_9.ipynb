```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
import os
from PIL import Image
```

```python

# Create output folder
os.makedirs("frames", exist_ok=True)
```

```python

# Generate multiple frames
prompt = "A mesmerizing animation of Delhi’s famous landmarks at twilight, featuring the Red Fort glowing under warm lights, India Gate standing tall with a patriotic aura, Qutub Minar reaching for the sky, and Lotus Temple reflecting on serene water. The scene transitions smoothly, blending historical grandeur with modern vibrancy, as the streets light up with rickshaws, street food stalls, and vibrant cultural life."
num_frames = 10  # Number of frames
```

```python

for i in range(num_frames):
    image = pipe(prompt).images[0]
    image.save(f"frames/frame_{i:03d}.png")  # Save frames
```

```python
import imageio
```

```python
frames = []
for i in range(num_frames):
    frame = Image.open(f"frames/frame_{i:03d}.png")
    frames.append(frame)
```

```python
# Save as a video
output_video = "animation.mp4"
imageio.mimsave(output_video, frames, fps=10)  # 10 FPS

print(f"Video saved as {output_video}")
```
